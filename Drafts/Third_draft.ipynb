```python
# Set up
import pandas as pd
import matplotlib.pyplot as plt
from googletrans import Translator
import re
from collections import Counter
import random
```

```python
#import df
df = pd.read_excel('interests_and_keywords_scholar_profile.xlsx')

### Convert the first letter into capitalize #####
# Convert the 'keywords' column to string type
df['keywords'] = df['keywords'].astype(str)

# Define a function to capitalize the first letter of each word
def capitalize_first_letter(text):
    # Split the string into words
    words = text.split()
    # Capitalize the first letter of each word
    capitalized_words = [word.capitalize() for word in words]
    # Join the words back into a single string
    return ' '.join(capitalized_words)

# Apply the function to the 'text' column
df['keywords'] = df['keywords'].apply(capitalize_first_letter)

# Select the elements based onf frecuency
#top_elements = element_counts[element_counts >= 4]

#top_elements.tail(60)
```

```python
def get_most_common_keywords(text, n=10):
    # Split the text into keywords
    words = text.split(',')
    # Count the frequencies of each word
    word_count = Counter(words)
    # Get the 'n' most common words
    most_common_words = [word for word, count in word_count.most_common(n)]
    # Join the most common words back into a single string
    return ','.join(most_common_words)

# Apply the function to each cell in the 'text' column
df['most_common_keywords'] = df['keywords'].apply(get_most_common_keywords)

# Print the result
#print(df['most_common_keywords'])
```

```python
############ / Just for visualize the data / ######################

###### Function to count word frequencies in a cell  #####
def word_frequencies(text):
    # Split the text into words
    words = text.split(',')
    # Count the frequencies of each word
    word_count = Counter(words)
    return word_count

# Apply the function to each cell in the 'text' column
df['keywords_frequencies'] = df['most_common_keywords'].apply(word_frequencies)

# Print the result
#print(df['keywords_frequencies'])

# Function to plot word frequencies
def plot_word_frequencies(word_count, title):
    words, counts = zip(*word_count.items())
    plt.figure(figsize=(10, 6))
    plt.bar(words, counts)
    plt.title(title)
    plt.xlabel('KeyWords')
    plt.ylabel('Frequencies')
    plt.show()

# Plot the word frequencies for each row
for index, row in df.iterrows():
    plot_word_frequencies(row['keywords_frequencies'], f"Frequencies for Row {index}")

```

```python
### Translate the keywords into english #########

def translate_keywords(code_snippet, origin= 'es', final = 'en'):
    # Initialize the translator
    translator = Translator()

    # Split the code snippet into lines
    lines = code_snippet.split(',')
    translated_lines = []

    for line in lines:
        # Detect the language of the line
        detected_lang = translator.detect(line).lang
        # Translate the line if it is in Spanish
        if detected_lang == origin:
            translated_line = translator.translate(line, src= origin, dest= final).text
        else:
            translated_line = line
        translated_lines.append(translated_line)

    # Join the translated lines back into a single string
    translated_code = ','.join(translated_lines)
    return translated_code

# Apply the function to the 'text' column
df['English'] = df['most_common_keywords'].apply(translate_keywords, origin = 'es', final = 'en')

df['Spanish'] = df['most_common_keywords'].apply(translate_keywords, origin = 'en', final = 'es')

```
